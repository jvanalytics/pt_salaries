```python
import pandas as pd


file_path=r'C:\Data\Coding\Python\pt_salaries\data\posts2024-08-16_18-43-21.json'


df = pd.read_json(file_path)
df['date'] = pd.to_datetime(df['created_utc'], unit='s').dt.strftime('%Y-%m-%d')

df.drop('created_utc',axis=1,inplace=True)

df.head(10)
```

# string cleaning

```python
# clean all ** from strings
df['content'] = df['content'].str.replace(r'\\\~|\~|\*\*', '', regex=True)
```

# clean dataframe Column creation using regex

```python


cleaned_df=df


cleaned_df['person_age'] = cleaned_df['content'].str.extract(r"Idade:.{0,5}(\d{2}).*")
cleaned_df['person_gender'] = cleaned_df['content'].str.extract(r"Idade:.+\((M|F)\)")
cleaned_df['person_education'] = cleaned_df['content'].str.extract(r"Formação acad.mic.*\:(.*)")
cleaned_df['person_work_experience'] = cleaned_df['content'].str.extract(r"Experi.ncia profission.*\:(.*)")
cleaned_df['person_marital_status'] = cleaned_df['content'].str.extract(r"Estado civil*\:(.*)")
cleaned_df['person_dependent_children'] = cleaned_df['content'].str.extract(r"dependentes.filhos*\:.*(\d+).*")

cleaned_df['company_industry'] = cleaned_df['content'].str.extract(r"Sector.Ind.stria.*\:(.*)")
cleaned_df['company_location'] = cleaned_df['content'].str.extract(r"Cidade.Regi.o.*egador*\:(.*)")
cleaned_df['company_headcount'] = cleaned_df['content'].str.extract(r"N.mero.*alhadores.*\:(.*)")
cleaned_df['company_international'] = cleaned_df['content'].str.extract(r"Multinacional.*\?(.*)")


cleaned_df['job_current'] = cleaned_df['content'].str.extract(r"Cargo atual.*\:(.*)")
cleaned_df['job_total_years'] = cleaned_df['content'].str.extract(r"Anos.*sa atual.*\:(.*)")
cleaned_df['job_workhours'] = cleaned_df['content'].str.extract(r"Horas.*trabalho.*\:(.*)")
cleaned_df['job_true_workhours'] = cleaned_df['content'].str.extract(r"M.dia.*semana.*ordinárias.*\:(.*)")
cleaned_df['job_shifts'] = cleaned_df['content'].str.extract(r"Trabalho.*turnos.*\:(.*)")
cleaned_df['job_24h'] = cleaned_df['content'].str.extract(r"Servi.*perman.*:(.*)")
cleaned_df['job_annual_holidays'] = cleaned_df['content'].str.extract(r"Dias.*ano.*:(.*)")


cleaned_df['conditions_annual_gross_income'] = cleaned_df['content'].str.extract(r"Sal.rio bruto.*\:(.*)")
cleaned_df['conditions_net_income'] = cleaned_df['content'].str.extract(r"Sal.rio.*l.q.*dio.*\:(.*)")
cleaned_df['conditions_liquid_benefits'] = cleaned_df['content'].str.extract(r"Compensa.*quidas.*\:(.*)")
cleaned_df['conditions_income_parcels'] = cleaned_df['content'].str.extract(r"m.s.*Duod.cimos.*\:(.*)")
cleaned_df['conditions_food_income'] = cleaned_df['content'].str.extract(r"Vales.*refei.*\:(.*)")
cleaned_df['conditions_company_car_and_fuel'] = cleaned_df['content'].str.extract(r"Carro .*empresa.*comb.*\:(.*)")
cleaned_df['conditions_insurance'] = cleaned_df['content'].str.extract(r"Seguro.*empregador.*\:(.*)")
cleaned_df['conditions_stock_bonus_other'] = cleaned_df['content'].str.extract(r"Outros.*compra.*es.*\:(.*)")

cleaned_df['mobility_job_location'] = cleaned_df['content'].str.extract(r"Cidade.*trabalho.*\:(.*)")
cleaned_df['mobility_travel_method'] = cleaned_df['content'].str.extract(r"Como.*deslocas.*\?(.*)")
cleaned_df['mobility_travel_compensation'] = cleaned_df['content'].str.extract(r"Como.*casa.trabalho.*\?(.*)")
cleaned_df['mobility_remote_work_days'] = cleaned_df['content'].str.extract(r"Dias.*teletrabalho.*\:(.*)")

cleaned_df['other_day_off_planning'] = cleaned_df['content'].str.extract(r"Com.*dia.*folga.*\?(.*)")
cleaned_df['other_stress_level'] = cleaned_df['content'].str.extract(r"O.*stressante.*\?(.*)")
cleaned_df['other_upskilling'] = cleaned_df['content'].str.extract(r"Possibilidades.*forma.*\:(.*)")
cleaned_df['other_job_leadership'] = cleaned_df['content'].str.extract(r"Respons.*projetos.*\:(.*)")



cleaned_df.drop(['author','content'],axis=1,inplace=True)

cleaned_df.head(10)
```
